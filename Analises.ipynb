## Funções para gráficos

```python
# Gráfico de barras vertical padrão
def grafico_barras_vertical(df, x_col, y_col, titulo, bar_color, fmt="{:.2f}"):

    # Parâmetros de cor
    bg_color  = "#F0F2F6"   # cor de fundo do gráfico
    # bar_color = "#3b82f6"   # cor das barras

    # Figura/Axes
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor(bg_color)
    ax.set_facecolor(bg_color)

    # Barras Seaborn
    sns.barplot(
        data=df,
        y=y_col,
        x=x_col,
        color=bar_color,
        ax=ax
    )

    # Adicionar os rótulos de dados com formatação
    for container in ax.containers:
        ax.bar_label(container, fmt=fmt, padding=5)

    # Título e rótulos
    ax.set_title(titulo, fontsize=16)
    ax.set_xlabel("")
    ax.set_ylabel("")

    # Remover valores do eixo X (ticks/labels)
    ax.set_xticks([])

    # Remover “quadrado” (spines)
    for spine in ["top", "right", "left", "bottom"]:
        ax.spines[spine].set_visible(False)

    plt.show()
```

```python
def grafico_barras_vertical_2(df, x_col, y_col, titulo, fmt="{:.2f}", hue_col=None, palette=None):

    # Parâmetros de cor
    bg_color  = "#F0F2F6"   # cor de fundo do gráfico

    # Figura/Axes
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor(bg_color)
    ax.set_facecolor(bg_color)

    # Barras Seaborn
    sns.barplot(
        data=df,
        y=y_col,
        x=x_col,
        hue=hue_col,  # Usar a coluna `hue` para a cor
        palette=palette, # Usar a paleta de cores personalizada
        ax=ax
    )

    # Adicionar os rótulos de dados com formatação
    for container in ax.containers:
        ax.bar_label(container, fmt=fmt, padding=5)

    # Título e rótulos
    ax.set_title(titulo, fontsize=16)
    ax.set_xlabel("")
    ax.set_ylabel("")

    # Remover valores do eixo X (ticks/labels)
    ax.set_xticks([])

    # Remover “quadrado” (spines)
    for spine in ["top", "right", "left", "bottom"]:
        ax.spines[spine].set_visible(False)

    plt.show()
```

```python
#Função para estatísticas e comparação
def analise_comparativa(escola_valor, serie_rede):
    media = serie_rede.mean()
    desvio = serie_rede.std()
    n = len(serie_rede)
    erro = desvio / np.sqrt(n)
    ic = stats.t.interval(0.95, df=n-1, loc=media, scale=erro)
    diferenca = escola_valor - media
    return {
        "Valor Escola": escola_valor,
        "Média Rede": media,
        "Desvio Padrão": desvio,
        "N": n,
        "IC 95%": ic,
        "Diferença": diferenca
    }

#Comparações
comparacao_tfp_estadual = analise_comparativa(vl_d_tfp, vl_d_estadual)
comparacao_tfp_municipal = analise_comparativa(vl_d_tfp, vl_d_municipal)

comparacao_tfp_estadual, comparacao_tfp_municipal
```
